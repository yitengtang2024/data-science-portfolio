```python
from sklearn.svm import SVR
import numpy as np
import matplotlib.pyplot as plt

# 生成一些示例数据
X = np.sort(5 * np.random.rand(40, 1), axis=0)
y = np.sin(X).ravel()

# 添加一些噪声
y[::5] += 3 * (0.5 - np.random.rand(8))

# 训练SVR模型
svr_rbf = SVR(kernel='rbf', C=100, epsilon=0.1)
svr_rbf.fit(X, y)
```

```python
# 生成测试数据
X_test = np.linspace(0, 5, 100).reshape(-1, 1)
y_pred = svr_rbf.predict(X_test)

# 画图
plt.scatter(X, y, color='darkorange', label='data')  # 原始数据点
plt.plot(X_test, y_pred, color='navy', lw=2, label='SVR prediction')  # SVR曲线
plt.fill_between(X_test.ravel(), 
                 y_pred - 0.1, 
                 y_pred + 0.1, 
                 color='lightblue', 
                 alpha=0.4, 
                 label='epsilon-tube')  # ε-tube范围

plt.xlabel('data')
plt.ylabel('target')
plt.title('Support Vector Regression (SVR)')
plt.legend()
plt.show()
```
