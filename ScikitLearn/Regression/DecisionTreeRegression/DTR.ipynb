```python
# 例子：面积（平方米） vs 房价（万美元）
X = np.array([[30], [40], [50], [60], [70], [80], [90], [100], [110], [120]])
y = np.array([100, 150, 200, 250, 300, 330, 360, 400, 430, 460])
```

```python
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
import numpy as np

# 训练决策树
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X, y)

# 创建细密的输入数据
X_grid = np.arange(X.min(), X.max(), 0.1).reshape(-1, 1)

# 预测
y_pred = regressor.predict(X_grid)

# 可视化
plt.scatter(X, y, color='red', label='Real Data')
plt.plot(X_grid, y_pred, color='blue', label='Prediction')
plt.title('Decision Tree Regression (Area vs Price)')
plt.xlabel('House Area (sqm)')
plt.ylabel('House Price ($10k)')
plt.legend()
plt.show()
```
