# Polynomial Regression

## Import necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import seaborn as sns
```

## Import data

```python
data = pd.read_csv("Position_Salaries.csv")
X = data.iloc[:, 1:2].values
y = data.iloc[:, -1].values
```

## Plot the data

```python
sns.scatterplot(x = X.flatten(), y = y)
plt.title("Position vs Salaries")
plt.xlabel("Position level")
plt.ylabel("Salaries")
plt.grid(True)
plt.show()
```

## Split data to train and test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Linear Model Fitting

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
sns.scatterplot(x = X_train.flatten(), y = y_train, color ='blue')
sns.lineplot(x = X_train.flatten(), y = reg.predict(X_train), color = 'red')
plt.title("Linear Model prediction")
plt.xlabel("Position level")
plt.ylabel("Salaries")
plt.grid(True)
plt.show()
```

## Polynomial model fitting

```python
poly = PolynomialFeatures(degree = 4)
poly.fit(X_train, y_train)
```

```python
X_train_poly = poly.fit_transform(X_train)
```

```python
model = LinearRegression()
model.fit(X_train_poly, y_train)
```

```python
X_plot = np.linspace(min(X),max(X),100).reshape(-1,1)
```

```python
X_plot_poly = poly.transform(X_plot)
```

```python
y_plot = model.predict(X_plot_poly)
```

```python
plt.scatter(X_train, y_train, color='blue',label="Training data")
plt.plot(X_plot, y_plot, color ='red', label="Polynomial Fit")
plt.title("Position vs Salaries(Poly model)")
plt.xlabel("Position level")
plt.ylabel("Salaries")
plt.grid(True)
plt.show()
```

## Predict a Number

```python
model.predict(poly.fit_transform([[6.5]]))
```
